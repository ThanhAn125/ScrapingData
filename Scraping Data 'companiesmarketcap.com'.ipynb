```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://companiesmarketcap.com/largest-companies-by-revenue/'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
table = soup.find_all('table')[0]
```

```python
title_all = table.find_all('th')
```

```python
title = [title.text.strip() for title in title_all][1:]
```

```python
df = pd.DataFrame(columns = title)
df.insert(3,'Unit','')
total_title = len(df.columns)
```

```python
table.find_all('tr').clear()
```

```python
data_row = table.find_all('td')
x = [data.text.strip() for data in data_row]
for i in x:
    if i == 'Close Ad X':
        x.remove(i)
start = 0
end = total_title+1
step = total_title
while end <= len(x) +1 :
    data = ' '.join(x[start:end]).strip()
    data_list = data.split()
    list1=[]
    for stt,data1 in enumerate(data_list[:len(data_list)-2]):
        if any(char.isalpha() for char in data1):
            list1.append(stt)
    change = list1[:len(list1)-1]
    data_list[min(change):max(change)+1] = [' '.join(data_list[min(change):max(change)+1])]
    if len(data_list) > 8:
        del(data_list[7])
    print(data_list)
    length = len(df)
    df.loc[length] = data_list
    start += step
    end += step
                
```

```python
df = df.drop(columns=['Price (30 days)'])
```
